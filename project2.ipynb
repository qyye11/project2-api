## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#save and print the text content of a page with all tags removed
url = "https://en.wikipedia.org/wiki/The_Game_Award_for_Game_of_the_Year"

response = requests.get(url)
soup = bs4.BeautifulSoup(response.text, 'html.parser')
text_content = soup.get_text()
print(text_content)
```

```python
#download an image with beautifulsoup and save it in this repository
from urllib.parse import urljoin

infobox = soup.find("table", {"class": "infobox"})

image_tag = infobox.find("img") if infobox else None

if image_tag:
    img_src = image_tag["src"]

    full_img_url = urljoin(url, img_src)
    print("Full image URL:", full_img_url)

    img_data = requests.get(full_img_url).content

    with open("the_game_award_infobox.jpg", "wb") as f:
        f.write(img_data)

    print("Image saved as the_game_award_infobox.jpg")
else:
    print("Could not find an infobox image on the page.")
```

```python
#find all the links in a page with beautifulsoup
all_links = soup.find_all("a")
print(f"Found {len(all_links)} total <a> tags.")

counter = 0
for link in all_links:
    href = link.get("href")
    if href:
        print(f"Link #{counter+1}: {href[:100]}")
        counter += 1
        if counter >= 10:
            break
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags

excerpts = []

def get_script_excerpt(url):
    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            text = soup.get_text()
            text = " ".join(text.split())
            return text[:1000]
        else:
            return "Failed to retrieve"
    except Exception as e:
        return str(e)

for link in scripts["link"]:
    excerpts.append(get_script_excerpt(link))

scripts["script_excerpt"] = excerpts
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"
scripts.to_csv("pudding_texts.csv", index=False)
```

## Part 4: TMDB database

#### Browse the documentation at https://developer.themoviedb.org/reference/intro/getting-started. Create an account to authenticate

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 
import requests
import os

API_KEY = "25d1eb8310d0198a12629b5f25aec138"
BASE_URL = "https://api.themoviedb.org/3"
NOW_PLAYING_ENDPOINT = f"{BASE_URL}/movie/now_playing"

params = {
    "api_key": API_KEY,
    "language": "en-US",
    "page": 1
}

response = requests.get(NOW_PLAYING_ENDPOINT, params=params)
now_playing_data = response.json()

movies = now_playing_data.get("results", [])
print("Number of movies returned:", len(movies))

dataset = []
for movie in movies:
    dataset.append({
        "id": movie.get("id"),
        "title": movie.get("title"),
        "release_date": movie.get("release_date"),
        "overview": movie.get("overview"),
        "vote_average": movie.get("vote_average"),
        "popularity": movie.get("popularity"),
        "original_language": movie.get("original_language"),
        "poster_path": movie.get("poster_path")
    })

for item in dataset:
    print(item["title"], "-", item["release_date"])
```

```python
#download the movie posters for 10 of these movies and save them to this repository
# Create a folder to hold the poster images, if it doesn't exist
os.makedirs("posters", exist_ok=True)

for i, movie in enumerate(dataset[:10]):
    poster_path = movie.get("poster_path")
    if poster_path:
        poster_url = f"https://image.tmdb.org/t/p/w500{poster_path}"
        print(f"Downloading poster for {movie['title']} from {poster_url}")

        poster_response = requests.get(poster_url)

        poster_filename = f"posters/{movie['id']}.jpg"
        with open(poster_filename, "wb") as f:
            f.write(poster_response.content)
    else:
        print(f"No poster path for {movie['title']}")
```
